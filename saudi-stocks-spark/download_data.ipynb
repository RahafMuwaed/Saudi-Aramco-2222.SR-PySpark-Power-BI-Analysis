```python
import yfinance as yf
import pandas as pd

symbol = "2222.SR"

data = yf.download(
    symbol,
    start="2019-01-01",
    interval="1d",
    auto_adjust=False,      # نحددها صراحة عشان يروح التحذير
    group_by="column"       # أهم سطر: يرجّع الأعمدة كـ Open/High/... بدون MultiIndex
)

# تحويل الـ index إلى عمود Date
data = data.reset_index()

# الآن الأعمدة لازم تكون بالشكل:
# Date, Open, High, Low, Close, Adj Close, Volume
print(data.columns)
data.head()
```

```python
import pandas as pd

# تسطيح (تفليت) الأعمدة: نأخذ المستوى الأول فقط
data.columns = data.columns.get_level_values(0)

# تأكد إن الأسماء صارت عادية
print(data.columns)
data.head()
```

```python
data["Symbol"] = "2222.SR" 

data.to_csv("data/raw/2222.SR.csv", index=False)
```

```python
import yfinance as yf
import pandas as pd

symbols = [
    "2010.SR",  # SABIC
    "1120.SR",  # Al Rajhi Bank
    "1180.SR",  # SNB
    "7010.SR",  # stc
    "1211.SR",  # Ma'aden
    "5110.SR",  # Saudi Electricity
    "2050.SR",  # Savola
    "4003.SR",  # Extra
    "1810.SR",  # Seera
]

for symbol in symbols:
    print("Downloading", symbol, "...")
    df = yf.download(
        symbol,
        start="2019-01-01",
        interval="1d",
        auto_adjust=False
    )
    df = df.reset_index()

    # لو الأعمدة MultiIndex (زي اللي صار معاك) نفكها
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df["Symbol"] = symbol

    path = f"data/raw/{symbol}.csv"
    df.to_csv(path, index=False)
    print("Saved:", path, "rows:", len(df))

print("Done")
```
